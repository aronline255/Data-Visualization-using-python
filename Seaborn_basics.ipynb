# Count plot

```python
# setting the figure size:  'figure.figsize':(width,height)
sns.set(rc={'figure.figsize':(5,5)})

sns.countplot(x='Sex', data = df, hue= 'Sex')
plt.show()
```

```python
ax = sns.countplot(x='Sex', data = df, hue='Pclass')

sns.set(rc={'figure.figsize':(15,5)})

# to print values on graph:
for bars in ax.containers:
    ax.bar_label(bars)
    
plt.show()    
```

# Kernel Density Estimation (KDE) plot

A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions.

```python
sns.set(rc={'figure.figsize':(6,5)})

sns.kdeplot(x='Age', data=df, color= 'Green')

plt.show()
```

# Distribution plot

This function provides access to several approaches for visualizing the univariate or bivariate distribution of data, including subsets of data defined by semantic mapping and faceting across multiple subplots. The kind parameter selects the approach to use:

histplot() (with kind="hist"; the default)

kdeplot() (with kind="kde")

```python
sns.displot(x='Age', data=df, kde=True, bins=12, hue='Survived')

# bins represents the number of intervals or bars

plt.show()
```

# Heat map
A heat map (or heatmap) is a 2-dimensional data visualization technique that represents the 
magnitude of individual values within a dataset as a color. 

```python
df3 = df2.drop('Species', axis= 1)
```

```python
corr = df3.corr()
sns.heatmap(corr)

plt.show()
```

# Histogram using Pandas
pandas.DataFrame.hist function calls matplotlib.pyplot.hist() , on each series in the DataFrame, resulting in one histogram per column. 

```python
df2.hist(edgecolor='black',bins = 25, figsize= (12,6))
plt.show()
```

# Joint plot
A seaborn jointplot shows bivariate scatterplots and univariate histograms in the same figure.

```python
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=df2)
plt.show()
```

# Violin plot

The Seaborn.violinplot() method allows the user to draw a box plot along with a kernel density estimate. Each violin is drawn using a kernel density estimate of the underlying distribution.

```python
sns.set_style("whitegrid") 

sns.violinplot(x = 'day', y = 'total_bill', data = dfr, hue='day') 
plt.show()
```
